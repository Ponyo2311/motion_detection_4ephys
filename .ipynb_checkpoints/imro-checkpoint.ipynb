```python
from scipy.io import savemat

directory = path
animal = 'AN_W-674'
file = 'LongCol_1shank_{}.imro'.format(animal)

with open(directory+file, 'r') as f:
    imro_contents = f.read()

locations = imro_contents.split(')(')[1:]

num_of_chans = len(locations)
# del_ind = []

for ind,line in enumerate(locations):
    coord, bank = [int(x) for x in locations[ind].split(' ', 2)[0:2]]
    # if coord in [191, 575]:
    #     del_ind.append(ind)
    locations[ind] = [bank, coord, abs(ind - num_of_chans)]

# delete ref channels
# for x in sorted(del_ind, reverse=True): locations.pop(x)

# sort list for banks
locations.sort()

sorted_list = [x[1] for x in locations]
sorted_list_str = ' '.join([str(x) for x in sorted_list])

# for Neuroscope2 0-based chan numbers
sorted_list_ns2 = list(reversed(sorted_list))
sorted_list_ns2 = ' '.join([str(x) for x in sorted_list_ns2])

# for NeuroScope2-Matlab 1-based chan numbers
sorted_list_mat = [x[1]+1 for x in locations]
sorted_list_mat = list(reversed(sorted_list_mat))
sorted_list_mat_str = ' '.join([str(x) for x in sorted_list_mat])

# Bank of every chan
sorted_bank = [x[0] for x in locations]
sorted_bank = list(reversed(sorted_bank))

# %% save data in mat file
#savemat(directory+'chanOrder_{}.mat'.format(animal),
#        {'sorted_list':sorted_list_mat, 'bank':sorted_bank, 'animal':animal})
```
