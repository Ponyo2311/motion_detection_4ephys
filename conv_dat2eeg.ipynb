```python
import os
import glob
import numpy as np
```

```python
fileTree={"name":["Dax","Odo","Timon","Pumba"]}
fileTree["data_number"]=["data-116","data-117"]
```

```python
sessions={}
toConvert=[]
for i in fileTree["data_number"]:
    for j in fileTree["name"]:
        path=os.path.join("/media/",i,j)
        if os.path.exists(path):
            sessions[j]={"data_nb":i,
                         "rat_name":j,
                         "dat2convert":[f for f in glob.glob(path+"*/**/Neuropix-PXI-100.1/*.dat", 
                                                 recursive=True) if os.path.isfile(f)],
                         "dat2convert_2":[f for f in glob.glob(path+"*/**/Neuropix-PXI-100.3/*.dat", 
                                                 recursive=True) if os.path.isfile(f)]}
            toConvert.append(sessions[j]["dat2convert"])
            toConvert.append(sessions[j]["dat2convert_2"])
```

```python
toConvert=np.concatenate(toConvert)
```
