# reading xml with ElementTree


```python
#import ElementTree
import xml.etree.ElementTree as ET
```

```python
#read xml
tree = ET.parse("/media/data-101/2021-07-18_19-30-44/RecordNode101/experiment1/recording1/continuous/Neuropix-PXI-100.1/continuous.xml")
```

```python
#get parent tag
root = tree.getroot()
print(root)
```

```python
#print the attributes of the first tag from the parent
print(root[1])
```

```python
for i in range(len(root)):
    #print (root[i])
    for j in range(len(root[i])):
        #print(root[i][j])
        for z in range(len(root[i][j])):
            #print(root[i][j][z])
            if 'group' in str(root[i][j][z]):
                groups = root[i][j][z]
                for g in groups:
                    print(g)
                    print(g.text)
            # print(root[i][j][z].attrib)
```
